# Prediction on single Image 

```python
from tensorflow.keras.preprocessing import image
from keras.models import load_model
from PIL import Image
model = load_model("../input/model-apple-disease/Final_apple_leaf_desease_detector_ResNet50_V2.h5")
Labels = ['complex', 'frog_eye_leaf_spot', 'healthy', 'powdery_mildew', 'rust', 'scab']
```

```python
import matplotlib.pyplot as plt
test_image   = '../input/plant-pathology-2021-fgvc8/test_images/85f8cb619c66b863.jpg'
image_result = Image.open(test_image)
test_image= image.load_img(test_image,target_size=(256,256))
test_image= image.img_to_array(test_image)
test_image= test_image/255
test_image=np.expand_dims(test_image,axis=0) # to tell nn dimention is (1,256,256,3)
result=model.predict(test_image)
result1 = result
result = result.tolist()[0]
image_result = plt.imshow(image_result)
l =[]
for i in result:
    
    if i>=0.50:
        j = result.index(i)
        l.append(Labels[j])
plt.title(str(l),fontweight= "bold")
#     else:
#         plt.title(Labels[np.argmax(result)])
        
plt.show()
```
